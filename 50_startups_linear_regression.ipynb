```python
Profit_num=df.select_dtypes(include=['float64'])
Profit_num
```

```python
df.corr()['Profit']
```

```python
Profit_num.head()
```

```python
Profit_num.shape
```

```python
x=Profit_num.iloc[:,0:3].values
x
```

```python
y=Profit_num.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0)
```

```python
Profit_num.shape
```

```python
x.shape
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred =model.predict(x_test)
y_pred
```

```python
len(y_pred)
```

```python
y_test
```

```python
df1=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
df1
```

```python
df1.plot(figsize=(10,8))
plt.show()
```

```python
df1.plot(figsize=(10,8),kind='bar')
plt.show()
```

```python
import seaborn as sns
sns.regplot(x='Actual',y='Predicted',data=df1,color='blue')
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
